```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
```

```python
#1.데이터 로드
loader = PyMuPDFLoader('data/car.pdf')
docs = loader.load()
print('문서 페이지수:',len(docs))
```

```python
docs[-1]
```

```python
#데이터 분활
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
split_documents = text_splitter.split_documents(docs)
print('청크의수:',len(split_documents))
```

```python
#임베딩
embedding = OpenAIEmbeddings()
```

```python
#벡터DB 생성 및 저장
db = Chroma.from_documents(
    documents=split_documents,
    embedding=embedding,
    persist_directory='db/chroma_db',
    collection_name = 'car_manual'
    )
```

```python
db.get()
```

```python
#검색기
retriever = db.as_retriever()
```

```python
retriever.invoke('차대번호 있는곳?')
```

```python
#프롬프트 작성
prompt = PromptTemplate.from_template(
    """
        You are an assistant for question-answering tasks.
        Use the following pieces of retrieved context to answer the question.
        If you don't know the answer, just say that you don't know.
        Answer in korean.

        previous chat history:
        {chat_history}

        question:
        {question}

        context:
        {context}

        answer:
    """
)
```

```python
#LLM생성
llm = ChatOpenAI(model_name='gpt-4o-mini',temperature=0)
```

```python
#체인생성
from langchain_core.output_parsers import StrOutputParser
from operator import itemgetter

chain = (
    {'context': itemgetter('question') | retriever , 'question' : itemgetter('question') , 'chat_history': itemgetter('chat_history') }
    |prompt
    |llm
    |StrOutputParser()
)
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

history_memory = {}

def get_session_history(session_ids):
    if session_ids not in history_memory:
        history_memory[session_ids] = ChatMessageHistory()
    return history_memory[session_ids]

history_chain = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key = 'question',
    history_messages_key='chat_history'
)
```

```python
history_chain.invoke(
    {'question':'브레이크 패드 교체주기는?'},
    config = {'configurable':{'session_id':'user123'}}
)
```
